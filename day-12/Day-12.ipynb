```python
import pandas as pd

import statsmodels.api as sm

from statsmodels.formula.api import ols

dataset8=pd.read_excel("D:/AI & ML/Lets Upgrade AI-ML/Assignments/Day-12/Practice Dataset/ANCOVA1.xlsx",sheet_name=0)

dataset8.head()
```

```python
model1=ols('Sales~C(Promotion)',dataset8).fit()
oneway=sm.stats.anova_lm(model1,typ=2)
print(oneway)
```

```python
model2=ols('Sales~C(Promotion)*C(Coupon)',dataset8).fit()

twoway=sm.stats.anova_lm(model2,typ=2)
print(twoway)
```
